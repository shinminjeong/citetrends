```python
import os, sys, json
import matplotlib.pyplot as plt
import numpy as np
```

```python
field = "assembly_language"
files = ["50_10000"]

for filename in files:
    infile = open('{}/{}.txt'.format(field, filename), 'r')
    data = json.load(infile)
    
    plt.figure(figsize=(20,5))
    plt.title(filename)
    for year in data.keys():
        plt.plot(data[year]["cite"].keys(), data[year]["cite"].values(), label=year)
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
for filename in files:
    infile = open('{}/{}.txt'.format(field, filename), 'r')
    data = json.load(infile)
    
    plt.figure(figsize=(20,5))
    plt.title(filename)
    for year in data.keys():
        numpub = data[year]["pub"] 
        if numpub == 0:
            continue
        plt.plot(data[year]["cite"].keys(), [v/numpub for v in data[year]["cite"].values()], label=year)
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
