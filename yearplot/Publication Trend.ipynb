```python
import os, sys, json
import matplotlib.pyplot as plt
import numpy as np
```

```python
fields = ["compiler", "deep_learning", "assembly_language", "flash_memory"]

data = {}
for field in fields:
    infile = open('{}/stat.txt'.format(field), 'r')
    data[field] = json.load(infile)
    
plt.figure(figsize=(15,5))
plt.title("Number of paper in citation bin")
for f in data.keys():
    plt.plot(data[f].keys(), [p["numpaper"] for p in data[f].values()], label=f)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
plt.figure(figsize=(15,5))
plt.title("Number of paper in citation bin")
for f in data.keys():
    plt.plot(data[f].keys(), np.log([p["numpaper"] for p in data[f].values()]), label=f)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
