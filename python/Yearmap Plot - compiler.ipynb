```python
import os, sys, json
import matplotlib.pyplot as plt
import numpy as np
```

```python
field = "compiler"
files = ["50_100", "100_150", "150_200", "200_10000"]

for filename in files:
    infile = open('{}/{}.txt'.format(field, filename), 'r')
    data = json.load(infile)
    
    plt.figure(figsize=(20,5))
    plt.title(filename)
    for year in data.keys():
        plt.plot(data[year].keys(), data[year].values(), label=year)
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
for filename in files:
    infile = open('{}/{}.txt'.format(field, filename), 'r')
    data = json.load(infile)
    
    plt.figure(figsize=(20,5))
    plt.title(filename)
    for year in data.keys():
        plt.plot(data[year].keys(), np.log(list(data[year].values())), label=year)
    plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
