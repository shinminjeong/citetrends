```python
import os, sys, json
import matplotlib.pyplot as plt
import numpy as np
```

```python
field = "nobel_winners"
winners = {
    "frank wilczek":["1973", "2004"],
    "brian p schmidt":["1998", "2011"],
    "george f smoot":["1992", "2006"],
    "yoichiro nambu":["1960", "2008"],
    "peter w higgs":["1964", "2012"]
}

for name in winners.keys():
    infile = open('{}/{}.txt'.format(field, name), 'r')
    data = json.load(infile)
    
    fig, ax = plt.subplots(figsize=(20,5))
    fig.suptitle(name, fontsize=20)
    for year in data.keys():
        ax.plot(data[year]["cite"].keys(), data[year]["cite"].values(), label=year)
        ax.plot([winners[name][0], winners[name][0]], [0, max(data[year]["cite"].values())], color='b', linestyle='--', linewidth=2)
        ax.plot([winners[name][1], winners[name][1]], [0, max(data[year]["cite"].values())], color='k', linestyle='--', linewidth=2)
#     ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    ax.grid(True)
    ax.xaxis.set_major_locator(plt.LinearLocator())
```
